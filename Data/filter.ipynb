## edges.csv

```python
'''
Drug - Gene: (total 9)
    DRUG_ACTIVATION_GENE = 'drug_activation_protein'
    DRUG_BINDING_GENE = 'drug_binding_protein'
    DRUG_CATALYSIS_GENE = 'drug_catalysis_protein'
    DRUG_EXPRESSION_GENE = 'drug_expression_protein'
    DRUG_INHIBITION_GENE = 'drug_inhibition_protein'
    DRUG_BINDACT_GENE = 'drug_bindingAndActivating_protein'
    DRUG_BINDINH_GENE = 'drug_bindingAndInhibiting_protein'
    DRUG_REACTION_GENE = 'drug_reaction_protein'
    DRUG_PREDBIND_GENE = 'drug_predictedBinding_protein'

Gene - Gene: (total 10)
    GENE_GENE = 'protein_interaction_protein'
    GENE_ACTIVATION_GENE = 'protein_activation_protein'
    GENE_BINDING_GENE = 'protein_binding_protein'
    GENE_CATALYSIS_GENE = 'protein_catalysis_protein'
    GENE_EXPRESSION_GENE = 'protein_expression_protein'
    GENE_INHIBITION_GENE = 'protein_inhibition_protein'
    GENE_BINDACT_GENE = 'protein_bindingAndActivating_protein'
    GENE_BINDINH_GENE = 'protein_bindingAndInhibiting_protein'
    GENE_REACTION_GENE = 'protein_reaction_protein'
    GENE_PTMOD_GENE = 'protein_postTranslationalModifications_protein'

Gene - Disease: (total 1)
    GENE_DIS = 'gene_interaction_disease

Disease - Drug: (total 1)
    DIS_DRUG = 'disease_indication_drug
'''
```

```python
input_file = r'..\HQ_DIR\graph_files\edges.csv'
output_file = r'..\HQ_DIR\graph_files\edges_relation_4type.csv'
```

```python
filter_values = {
    'DRUG_ACTIVATION_GENE',
    'DRUG_BINDING_GENE',
    'DRUG_CATALYSIS_GENE',
    'DRUG_EXPRESSION_GENE',
    'DRUG_INHIBITION_GENE',
    'DRUG_BINDACT_GENE',
    'DRUG_BINDINH_GENE',
    'DRUG_REACTION_GENE',
    'DRUG_PREDBIND_GENE',
    'GENE_GENE',
    'GENE_ACTIVATION_GENE',
    'GENE_BINDING_GENE',
    'GENE_CATALYSIS_GENE',
    'GENE_EXPRESSION_GENE',
    'GENE_INHIBITION_GENE',
    'GENE_BINDACT_GENE',
    'GENE_BINDINH_GENE',
    'GENE_REACTION_GENE',
    'GENE_PTMOD_GENE',
    'GENE_DIS',
    'DIS_DRUG'
}

def filter_edges(input_file, output_file, edge_types):
    with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        reader = csv.reader(infile, delimiter='\t')
        writer = csv.writer(outfile, delimiter='\t')
        
        for row in reader:
            if row[1] in edge_types:
                writer.writerow(row)

filter_edges(input_file, output_file, filter_values)
```

```python
def count_rows(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.reader(file, delimiter='\t')
        return sum(1 for row in reader)

original_count = count_rows(input_file)
filtered_count = count_rows(output_file)

print(f"edges.csv 파일의 항목 개수: {original_count}")
print(f"edges_relation_4type.csv 파일의 항목 개수: {filtered_count}")
```
