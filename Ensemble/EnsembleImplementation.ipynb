```python
def clean_entity(e):

    entity2id_path = r'all\entity2id.txt'
    entity2id = {}
    with open(entity2id_path, 'r') as f:
        next(f)
        for line in f:
            line = line.strip()
            if not line:
                continue
            name, idx = line.split('\t')
            entity2id[int(idx)] = name

    entity_embedding = {}
    with open(e, 'r') as f:
        for line in f:
            entity, vector = line.strip().split(': ')
            entity_num=int(entity.split('_')[1])
            vector = eval(vector.strip())
            entity_embedding[entity_num] = vector
    embedding_df = pd.DataFrame.from_dict(entity_embedding, orient='index')
    embedding_df.columns = [f'dim_{i}' for i in range(embedding_df.shape[1])]
    embedding_df['name'] = embedding_df.index.map(entity2id)
    embedding_df = embedding_df[['name'] + [col for col in embedding_df.columns if col != 'name']]
    embedding_df = embedding_df.sort_index()

    return embedding_df


def clean_relation(r):

    relation2id_path = r'all\relation2id.txt'
    relation2id = {}
    with open(relation2id_path, 'r') as f:
        next(f)
        for line in f:
            line = line.strip()
            if not line:
                continue
            name, idx = line.split('\t')
            relation2id[int(idx)] = name

    relation_embedding = {}
    with open(r, 'r') as f:
        for line in f:
            relation, vector = line.strip().split(': ')
            relation_num = int(relation.split('_')[1])
            vector = eval(vector.strip())
            relation_embedding[relation_num] = vector
    embedding_df = pd.DataFrame.from_dict(relation_embedding, orient='index')
    embedding_df.columns = [f'dim_{i}' for i in range(embedding_df.shape[1])]
    embedding_df['name'] = embedding_df.index.map(relation2id)
    embedding_df = embedding_df[['name'] + [col for col in embedding_df.columns if col != 'name']]
    embedding_df = embedding_df.sort_index()

    return embedding_df


def clean_norm(n):

    relation2id_path = r'all\relation2id.txt'
    relation2id = {}
    with open(relation2id_path, 'r') as f:
        next(f)
        for line in f:
            line = line.strip()
            if not line:
                continue
            name, idx = line.split('\t')
            relation2id[int(idx)] = name

    norm_embedding = {}
    with open(n, 'r') as f:
        for line in f:
            norm, vector = line.strip().split(': ')
            norm_num = int(norm.split('_')[2])
            vector = eval(vector.strip())
            norm_embedding[norm_num] = vector
    embedding_df = pd.DataFrame.from_dict(norm_embedding, orient='index')
    embedding_df.columns = [f'dim_{i}' for i in range(embedding_df.shape[1])]
    embedding_df['name'] = embedding_df.index.map(relation2id)
    embedding_df = embedding_df[['name'] + [col for col in embedding_df.columns if col != 'name']]
    embedding_df = embedding_df.sort_index()

    return embedding_df


def clean_matrix(m):
    import pandas as pd
    import ast

    relation2id_path = r'all\relation2id.txt'
    id2relation = {}
    with open(relation2id_path, 'r') as f:
        next(f)
        for line in f:
            line = line.strip()
            if not line:
                continue
            name, idx = line.split('\t')
            id2relation[int(idx)] = name  # key: int idx, value: relation name

    relation_matrix_df = {}
    with open(m, 'r') as f:
        current_relation = None
        current_matrix = []
        for line in f:
            line = line.strip()
            if not line:
                continue
            if line.startswith('Matrix_'):
                if current_relation is not None:
                    relation_name = id2relation[current_relation]
                    relation_matrix_df[relation_name] = pd.DataFrame(current_matrix)
                current_relation = int(line.split('_')[1].rstrip(':'))
                current_matrix = []
            else:
                current_matrix.append(ast.literal_eval(line))

        if current_relation is not None:
            relation_name = id2relation[current_relation]
            relation_matrix_df[relation_name] = pd.DataFrame(current_matrix)

    return relation_matrix_df
```

```python
id_cols = ['Drug', 'Gene']
relation_cols = [col for col in transe_dg_pred.columns if col not in id_cols]

transe_dg_pred_scaled = (transe_dg_pred[relation_cols] - params['transe_dg_min']) / (params['transe_dg_max'] - params['transe_dg_min'])
transh_dg_pred_scaled = (transh_dg_pred[relation_cols] - params['transh_dg_min']) / (params['transh_dg_max'] - params['transh_dg_min'])
transr_dg_pred_scaled = (transr_dg_pred[relation_cols] - params['transr_dg_min']) / (params['transr_dg_max'] - params['transr_dg_min'])
complex_dg_pred_scaled = (complex_dg_pred[relation_cols] - params['complex_dg_min']) / (params['complex_dg_max'] - params['complex_dg_min'])
rotate_dg_pred_scaled = (rotate_dg_pred[relation_cols] - params['rotate_dg_min']) / (params['rotate_dg_max'] - params['rotate_dg_min'])
rgcn_dg_pred_scaled = (rgcn_dg_pred[relation_cols] - params['rgcn_dg_min']) / (params['rgcn_dg_max'] - params['rgcn_dg_min'])
hitter_dg_pred_scaled = (hitter_dg_pred[relation_cols] - params['hitter_dg_min']) / (params['hitter_dg_max'] - params['hitter_dg_min'])

weighted_transe  = transe_dg_pred_scaled[relation_cols] * weights['transe']
weighted_transh  = transh_dg_pred_scaled[relation_cols] * weights['transh']
weighted_transr  = transr_dg_pred_scaled[relation_cols] * weights['transr']
weighted_complex = complex_dg_pred_scaled[relation_cols] * weights['complex']
weighted_rotate  = rotate_dg_pred_scaled[relation_cols] * weights['rotate']
weighted_rgcn  = rgcn_dg_pred_scaled[relation_cols] * weights['rgcn']
weighted_hitter  = hitter_dg_pred_scaled[relation_cols] * weights['hitter']

weighted_sum = (
    weighted_transe +
    weighted_transh +
    weighted_transr +
    weighted_complex +
    weighted_rotate + 
    weighted_rgcn + 
    weighted_hitter
)

total_weight = sum(weights.values())
weighted_avg = weighted_sum / total_weight

dg_pred = transe_dg_pred[id_cols].copy()
dg_pred = pd.concat([dg_pred, weighted_avg], axis=1)
```
